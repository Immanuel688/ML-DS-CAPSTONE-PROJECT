```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
dataset= pd.read_csv("Preprocessed_dataset.csv")
dataset1=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
indep_x= dataset1.drop('Conversion',axis=1)
dep_y=dataset1['Conversion']
```

```python
indep_x
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif
def selectkbest(indep_x,dep_y,n):
        test = SelectKBest(score_func=mutual_info_classif, k=n)
        fit1= test.fit(indep_x,dep_y)
        featurename= fit1.get_support(indices=True)
        selected_features=dataset1.columns[featurename]
        features=selected_features.to_list()
        print(f"The {n} features are",features)
        selectk_features = fit1.transform(indep_x)
        return selectk_features
```

```python
selectk_features=selectkbest(indep_x,dep_y,12)
selectk_features
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
def split_scalar(indep_x,dep_y):
        X_train, X_test, y_train, y_test = train_test_split(indep_x, dep_y, test_size = 0.2, random_state = 42)   
        return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test=split_scalar(selectk_features,dep_y)
X_train
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
LRC=LogisticRegression(random_state = 0,)
LRC.fit(X_resampled, y_resampled)
LRC_probs =LRC.predict_proba(X_test)[:, 1]
LRC_probs
```

```python
from xgboost import XGBClassifier
#from collections import Counter
#counter = Counter(y_train)  # Count class occurrences
#ratio = counter[0] / counter[1]
# Meta-Model (different from base models)
param_grid_XGB={'n_estimators': [100],'max_depth': [3, 6],'learning_rate': [0.01, 0.1],'subsample': [ 0.8],'colsample_bytree': [0.7,1.0],'gamma': [1]}
grid_XGB =  GridSearchCV(XGBClassifier(), param_grid_XGB, cv=5)
grid_XGB.fit(X_resampled,y_resampled)
XGB_probs = grid_XGB.predict_proba(X_test)[:, 1]
XGB_probs
```

```python
import numpy as np
from sklearn.metrics import f1_score, precision_recall_curve

# Initialize variables
best_f1 = 0
best_w1 = 0
best_w2 = 0
best_threshold = 0.5  # Default threshold

# Loop over possible weights for LRC (w1) from 0.1 to 1.0
for w1 in np.arange(0.1, 1.1, 0.1):  # Include 1.0
    w2 = 1 - w1  # Complementary weight for XGB
    blended_probs = (w1 * LRC_probs) + (w2 * XGB_probs)  # Blend probabilities
    
    # Find the best threshold for the current weights
    precision, recall, thresholds = precision_recall_curve(y_test, blended_probs)
    for threshold in thresholds:
        blended_preds = (blended_probs >= threshold).astype(int)
        f1 = f1_score(y_test, blended_preds)
        if f1 > best_f1:
            best_f1 = f1
            best_w1 = w1
            best_w2 = w2
            best_threshold = threshold

print(f"Best weights: LRC = {best_w1}, XGB = {best_w2}, Best threshold: {best_threshold}, Best F1-score: {best_f1}")
```

```python
 #Blend models using the best weights
final_probs = (best_w1 * LRC_probs) + (best_w2 * XGB_probs)

# Apply the best threshold to make final predictions
final_pred = (final_probs >= best_threshold).astype(int)

# Evaluate the final predictions
from sklearn.metrics import classification_report, confusion_matrix

print("Classification Report:")
print(classification_report(y_test, final_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, final_pred))
```
