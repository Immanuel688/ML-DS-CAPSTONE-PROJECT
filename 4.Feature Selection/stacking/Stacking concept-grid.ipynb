```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
dataset= pd.read_csv("Preprocessed_dataset.csv")
dataset1=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
indep_x= dataset1.drop('Conversion',axis=1)
dep_y=dataset1['Conversion']
```

```python
indep_x
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif
def selectkbest(indep_x,dep_y,n):
        test = SelectKBest(score_func=mutual_info_classif, k=n)
        fit1= test.fit(indep_x,dep_y)
        featurename= fit1.get_support(indices=True)
        selected_features=dataset1.columns[featurename]
        features=selected_features.to_list()
        print(f"The {n} features are",features)
        selectk_features = fit1.transform(indep_x)
        return selectk_features
```

```python
selectk_features=selectkbest(indep_x,dep_y,12)
selectk_features
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(indep_x, dep_y, test_size = 0.2, random_state = 42)   
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train= sc.fit_transorm(X_train)
X_test= sc.transfrom(X_test)
```

```python
from xgboost import XGBClassifier
from collections import Counter
counter = Counter(y_train)  # Count class occurrences
ratio = counter[0] / counter[1]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_LRC= {'C': [0.01, 0.1, 1], 'penalty': ['l2'], 'solver': ['sag'],'max_iter': [1200],'random_state':[42]}
#param_grid_DT={'criterion':["gini", "entropy", "log_loss"], 'splitter': ["best", "random"],'max_features': [None,"sqrt", "log2"], 'random_state':[42]}
param_grid_XGB={'n_estimators': [100],'max_depth': [3, 6],'learning_rate': [0.01, 0.1],'subsample': [ 0.8],'colsample_bytree': [0.7,1.0],'gamma': [1]}
```

```python
LRC=GridSearchCV(LogisticRegression(class_weight='balanced'), param_grid_LRC, cv=5, scoring='f1', n_jobs=-1,refit=True)
#DT=GridSearchCV(DecisionTreeClassifier(class_weight='balanced'), param_grid_DT, refit= True, verbose=3, n_jobs=-1, scoring= 'f1' )
XGB= GridSearchCV(XGBClassifier(scale_pos_weight=ratio), param_grid_XGB, refit= True, verbose=3, n_jobs=-1, scoring= 'f1' )
```

```python
LRC.fit(X_train, y_train )
#DT.fit(X_train, y_train)
XGB.fit(X_train, y_train )
```

```python
base_models = [
    ('XGB', XGB.best_estimator_),
    ('LRC', LRC.best_estimator_),
    #('DT', DT.best_estimator_)
]
```

```python
base_models
```

```python
param_grid_meta={'final_estimator__criterion':["gini", "entropy", "log_loss"],'final_estimator__max_features': [None,"sqrt", "log2"], 'final_estimator__n_estimators':[10,50,100],'final_estimator__random_state':[42]}
```

```python
# Meta-Model (different from base models)
meta_model =  RandomForestClassifier(class_weight='balanced')
```

```python
stacking_clf = StackingClassifier(estimators=base_models, final_estimator=meta_model)
```

```python
stacking_clf 
```

```python
grid_meta = GridSearchCV(stacking_clf, param_grid_meta, cv=3, scoring='f1', n_jobs=-1)
```

```python
grid_meta 
```

```python
grid_meta.fit(X_train,y_train)
```

```python
y_pred=grid_meta.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
report=classification_report(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
```
