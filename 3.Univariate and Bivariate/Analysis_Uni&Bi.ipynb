```python
quan = []
quan.extend(pre_dataset.describe().columns)
```

```python
quan
```

```python
analysis= pd.DataFrame(index=['mean', 'median','mode','Q1:25th', 'Q2:50th','Q3:75th','Q4:100th','min','max','IQR','1.5 Rule','Lesser range','Greater range'], columns=quan)
```

```python
analysis
```

```python
for columnName in quan:
    analysis.loc['mean',columnName]= pre_dataset[columnName].mean()
    analysis.loc['median',columnName]=pre_dataset[columnName].median()
    analysis.loc['mode',columnName]=pre_dataset[columnName].mode()[0]
    analysis.loc['Q1:25th',columnName]=np.percentile(pre_dataset[columnName],25)
    analysis.loc['Q2:50th',columnName]=np.percentile(pre_dataset[columnName],50)
    analysis.loc['Q3:75th',columnName]=np.percentile(pre_dataset[columnName],75)
    analysis.loc['Q4:100th',columnName]=np.percentile(pre_dataset[columnName],100)
    analysis.loc['min',columnName]=pre_dataset[columnName].min()
    analysis.loc['max',columnName]=pre_dataset[columnName].max()
    analysis.loc['IQR',columnName]=analysis.loc['Q3:75th',columnName]-analysis.loc['Q1:25th',columnName]
    analysis.loc['1.5 Rule',columnName]= 1.5*analysis.loc['IQR',columnName]
    analysis.loc['Lesser range', columnName]= analysis.loc['Q1:25th',columnName]- analysis.loc['1.5 Rule',columnName]
    analysis.loc['Greater range', columnName]= analysis.loc['Q3:75th',columnName]+ analysis.loc['1.5 Rule',columnName]
```

```python
analysis
```

```python
# Checking for outliers
def find_outiers(analysis,quan):
    greater=[]
    lesser=[]
    for columnName in quan:
        if (analysis.loc['min',columnName]<analysis.loc['Lesser range', columnName]):
            lesser.append(columnName)
        if (analysis.loc['max',columnName]>analysis.loc['Greater range', columnName]): 
            greater.append(columnName)
    return greater, lesser
```

```python
# Frequency table
```

```python
def freqtable(columnName,pre_dataset):
    freqtable=pd.DataFrame(columns=['unique_values','frequency','relative_frequency','cumsum'])
    freqtable['unique_values']=pre_dataset[columnName].value_counts().index
    freqtable['frequency']=pre_dataset[columnName].value_counts().values
    freqtable['relative_frequency']= freqtable['frequency']/freqtable['frequency'].sum()
    freqtable['cumsum']=freqtable['relative_frequency'].cumsum()
    return freqtable
```

```python
# How many customers vistited the website more than 25 times
visits=freqtable('WebsiteVisits',pre_dataset)
visits[visits['unique_values']>25]['frequency'].sum()
```

```python
# How many customers clicked the link from the marketing email more than 5 times?
clicks=freqtable('EmailClicks',pre_dataset)
clicks[clicks['unique_values']>5]['frequency'].sum()
```
